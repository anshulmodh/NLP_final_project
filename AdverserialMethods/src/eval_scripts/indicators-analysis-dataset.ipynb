```python
import pandas as pd
import random
```

```python
word_counts_train = dict()
word_counts_val = dict()
joints = {"neutral_train": {}, "contradiction_train": {}, "entailment_train": {}, "neutral_val": {}, "contradiction_val": {}, "entailment_val": {}}
```

```python
def add_word(w, d):
    if w in d:
        d[w] += 1
    else:
        d[w] = 1
```

```python
# filename_train = "../data/snli_1.0/cl_snli_train_source_file"
# filename_val = "../data/snli_1.0/cl_snli_val_source_file"
# filename_train_labels = "../data/snli_1.0/cl_snli_train_lbl_file"
# filename_val_labels = "../data/snli_1.0/cl_snli_val_lbl_file"
# filename_train = "../data/sick/cl_sick_train_source_file"
# filename_val = "../data/sick/cl_sick_val_source_file"
# filename_train_labels = "../data/sick/cl_sick_train_lbl_file"
# filename_val_labels = "../data/sick/cl_sick_val_lbl_file"
filename_train = "../data/snli_1.0/cl_snli_train_source_file"
filename_val = "min0.9,scaled--samples.txt"
filename_train_labels = "../data/snli_1.0/cl_snli_train_lbl_file"
filename_val_labels = "min0.9,scaled--samples_labels.txt"
train_labels = open(filename_train_labels, 'r')
train = open(filename_train, 'r', encoding="utf-8")
if "multinli" in filename_val:
    val = open(filename_val, 'rb')
else:
    val = open(filename_val, 'r')
val_labels = open(filename_val_labels, 'r', encoding="utf-8")
min_conditional = 0.9
min_support = 20
```

```python
num_samples = 0

try:
    while True:
        label = next(train_labels).replace("\n", "")
        hypothesis = set(next(train).lower().split("|||")[1].replace("\n", "").split(" "))
        for word in hypothesis:
            add_word(word, word_counts_train)
            add_word(word, joints[label + "_train"])
        num_samples += 1
        if num_samples % 10000 == 0:
            print(num_samples, "completed")
except:
    pass
```

```python
num_samples = 0

try:
    while True:
        label = next(val_labels).replace("\n", "")
        if "multinli" in filename_val:
            try:
                hypothesis = set(next(val).decode(encoding='UTF-8').lower().split("|||")[1].replace("\n", "").split(" "))
            except UnicodeDecodeError:
                pass
        else:
            hypothesis = set(next(val).lower().split("|||")[1].replace("\n", "").split(" "))
        for word in hypothesis:
            add_word(word, word_counts_val)
            add_word(word, joints[label + "_val"])
        num_samples += 1
        if num_samples % 10000 == 0:
            print(num_samples, "completed")
except StopIteration:
    pass
```

```python
words = pd.DataFrame.from_dict(word_counts_train, orient='index').rename(columns={0: "count_train"})
words = words.join(pd.DataFrame.from_dict(word_counts_val, orient='index').rename(columns={0: "count_val"}), how="outer")
words.head()
```

```python
words = words.join(pd.DataFrame.from_dict(joints["neutral_train"], orient='index').rename(columns={0: "neutral_train"}), how="outer")
words = words.join(pd.DataFrame.from_dict(joints["contradiction_train"], orient='index').rename(columns={0: "contradiction_train"}), how="outer")
words = words.join(pd.DataFrame.from_dict(joints["entailment_train"], orient='index').rename(columns={0: "entailment_train"}), how="outer")
words = words.join(pd.DataFrame.from_dict(joints["neutral_val"], orient='index').rename(columns={0: "neutral_val"}), how="outer")
words = words.join(pd.DataFrame.from_dict(joints["contradiction_val"], orient='index').rename(columns={0: "contradiction_val"}), how="outer")
words = words.join(pd.DataFrame.from_dict(joints["entailment_val"], orient='index').rename(columns={0: "entailment_val"}), how="outer")
words.head()
```

```python
words = words.sort_values("count_train", ascending=False).reset_index().rename(columns={"index": "word"})
words.head()
```

```python
words["neutral_train"] = words["neutral_train"] / words["count_train"]
words["contradiction_train"] = words["contradiction_train"] / words["count_train"]
words["entailment_train"] = words["entailment_train"] / words["count_train"]
words["neutral_val"] = words["neutral_val"] / words["count_val"]
words["contradiction_val"] = words["contradiction_val"] / words["count_val"]
words["entailment_val"] = words["entailment_val"] / words["count_val"]
words.head()
```

```python
words = words[(words["neutral_train"] > min_conditional) | (words["contradiction_train"] > min_conditional) | (words["entailment_train"] > min_conditional)]
words.head()
```

```python
words = words[words['count_train'] > min_support]
words.head()
```
